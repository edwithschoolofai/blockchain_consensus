# Proof of work
![alt text](https://image.slidesharecdn.com/blockchainbtsym-161201162319/95/blockchain-btsym-16-11-638.jpg?cb=1480610107 "Logo Title Text 1")

![alt text](https://image.slidesharecdn.com/blockchainbtsym-161201162319/95/blockchain-btsym-16-12-638.jpg?cb=1480610107 "Logo Title Text 1")

```python
#간단한 Proof-of-Work 예제

import sys
import time
import hashlib 
from struct import unpack, pack

#timestamp, message and payload 들은 단지 “재료” 입니다
#여러분들이 네트워크로 보낼 것들이죠
#비트코인 마이닝에서처럼, 블록 데이터 같은 것들 입니다.

timestamp = str(time.time())
message = "this is a random message"
payload = timestamp + message

#nonce, guess, throttle, target은 작업 수행에 사용됩니다. 
#proof-of-work에서 가장 중요한 부분은
#클라이언트에게는 어려워야하며 
#시스템에서는 체크가 쉬워야한다는 점입니다.
#매번 추측할 때 마다 none만 변합니다. 
#정답일 가능성은 매우 낮기 때문에  
#이 시스템은 확률적으로 
#많은 추측 또는 작업이 수행되었음을 보증해줍니다.

nonce = 0
guess = 999999999999
throttle = 100000000
target = 2**64/ throttle

#
#매번 nonce는 1씩 증가 
#컴퓨터가 몇 번의 추측을 했는 지 효과적으로 카운팅합니다.
#그러나, 필수는 아닙니다.

#시스템이 작업을 체크할 때, 
#매번 추측한 것들을 다 추적할 필요는 없습니다. 
#단지 데이터와 함께 제출된 nonce만 보면 됩니다.


payloadHash = hashlib.sha512(payload).digest()

start = time.time()

#다음 세 줄의 코드가 proof-of-work 알고리즘입니다. 
#단순한 루프이며, 데이터를 SHA256 으로 두 번 해싱합니다. 
#처음 8바이트는 숫자로 변환됩니다.
#효과적으로, 이것이 컴퓨터가 추측을 하는 방법입니다.
while guess > target:
    nonce+1
    guess, = unpack('>Q', hashlib.sha512(hashlib.sha512(pack('>Q', nonce) +
                                                       payloadHash).digest()).digest())[0:8]
end = time.time()
#nonce는 작업이 수행되었음을 증명해주는 것입니다. 
#이 예제에서는, 여러분들의 CPU가 시도한 값을 의미하는데
#그 값은 타겟값 이하의 유효한 값을 찾기 전까지 시도를 의미합니다. 매번 추측값이 타겟값 이하가 될 확률은 동일하기 때문에 
#nonce를 생성하는 방법은 중요하지 않습니다. 하지만 단지 nonce를 증가시키는 연산은
#랜덤하게 숫자를 생성할 때보다 비용이 낮습니다. 네트워크로 작업이 제출되었을 때
#nonce는 정답을 검증하는데 사용됩니다.
print "%s:%s:%s:%s:%s:%s:%s" % (timestamp, message, nonce, guess, payload, target, end-start)
```

# Proof of Stake

![alt text](https://github.com/edwithschoolofai/blockchain_consensus/raw/master/POW-vs-POS.jpg "Logo Title Text 1")

- Proof of Stake (PoS)는 퍼블릭 블록체인의 합의 알고리즘의 한 종류로, 네트워크의 검증자의 경제적 지분에 좌우됩니다.
- proof of work (PoW) 기반의 퍼블릭 블록체인들(e.g. 비트코인이나, 현 시점의 이더리움)에서는, 알고리즘이 암호 퍼즐을 푼 참가자에게 보상을 줍니다. 이는 거래의 유효성을 검증하고 새로운 블록 생성에 사용됩니다. (i.e. 채굴). 
- PoS 기반의 퍼블릭 블록체인들(e.g. 다음 버전의 이더리움인 Casper), 여러 검증자들이 순번대로 다음 블록에 대해 제안과 투표를 하게 되며, 각 검증자들의 투표 가중치는 그들의 예금(i.e. 지분/stake)에 따라 좌우됩니다. 
- PoS의 눈에 띄는 장점은 보안, 중앙화의 리스크 감소, 효율적인 에너지 사용 등입니다.

일반적으로 PoS 알고리즘은 다음과 같습니다.

- 블록체인은 검증자들 그룹을 트랙킹합니다.
- 누구라도 특별한 타입의 거래를 전송하면 검증자가 될 수 있으며, 이때 그들의 코인은 예금으로 묶입니다.
- 새 블록에 생성의 동의는 현재의 모든 검증자들이 참여할 수 있는 합의 시스템으로 결정됩니다.
- 합의 알고리즘에 참여하는 검증자들에게 보상을 주는 방법들은 많은데 대표적인 두 가지 타입은: 체chain-based proof of stake와 BFT-style proof of stake입니다.

chain-based proof of stake에서는
- 알고리즘이 랜덤스럽게(pseudo-randomly) 검증자들을 각 타입 슬롯(eg. 매번 10초동안) 도중에 선택하게 됩니다.
- 그 검증자들에게 1개의 블록을 생성할 수 있는 권한을 줍니다.
- 이 블록은 이전 블록을 포인팅하고 있어야만 합니다. (보통은 이전에 가장 긴 체인의 마지막 블록을 포인팅합니다.), 시간이 지나면 대부분의 블록들이 계속 커지는 하나의 큰 체인 속으로 모이게 됩니다.

BFT-style proof of stake (비잔틴 폴트 톨러런트 byzantine fault tolerant)
- 검증자들은 랜덤하게 블록을 제안할 권한을 갖게 됩니다.
- 어떤 블록이 정식으로 체택될 지에 관한 동의는 멀티-라운드 과정을 거쳐 결정되며, 모든 검증자들은 매 라운드마다 특정 블록에 "투표" 합니다. 
- 이 과정의 끝에서는, 모든 (정직하고, 접속되어 있는) 검증자들이 주어진 블록이 체인의 일부가 될지 말지에 관한 영구적인 동의를 하게 됩니다.
- 블록들은 여전히 서로 연결되어 있을 수 있습니다.; 이 합의 알고리즘의 가장 큰 차별점은 블록 하나를 다룰 뿐이며, 이 블록이 속한 체인의 길이는 관련이 없습니다.

### Proof of Activity (PoW + PoS 하이브리드) - DeCred 코인

- 전통적인 방식의 proof-of-work을 합니다. (채굴자들이 암호 퍼즐을 풀기 위해 레이스를 합니다.)
- 채굴된 블록들은 아무 거래내역도 가지고 있지 않습니다. 따라서 레이스에서 우승한 블록은 단지 header와 채굴자의 보상을 위한 주소만 가지고 있게 됩니다.
- 그 후에, 시스템이 proof of stake로 전환됩니다.
- header 정보에 근거하여 랜덤한 검증자들이 선택되고, 새로운 블록을 승인하게 됩니다.
- 시스템 내의 더 많은 코인들을 소유하고 있는 검증자들가, 선택될 가능성이 더 높아집니다. 
- 껍데기에 불과한 블록은, 모든 검증자들이 승인함과 동시에 완전히 정식의 블록이 되게 됩니다.

### Proof of Burn

- 비싼 컴퓨터 장비에 돈을 쏟아 붓는 대신, 코인들을 돌려받지 못할 주소로 송금하여 '소각'합니다.
- 영영 돌아오지 못할 곳으로 코인을 보내면, 대신 시스템에서 랜덤하게 선택되어 평생 채굴할 수 있는 특권을 받게 됩니다.
- 시간이 지나면, 시스템 내의 여러분들의 지분은 줄어들게 되며, 결국 더 많은 코인을 소각하여 로또에 당첨될 확률을 높이고 싶어질 것입니다. (이 부분은 비트코인의 채굴을 흉내낸 것인데요, 해싱 파워를 유지하기 위해 최신 컴퓨팅 장비를 계속 더 사야됩니다.)

### 제가 가장 좋아하는 방식
- 고유 식별 증명(Proof of unique identity)
